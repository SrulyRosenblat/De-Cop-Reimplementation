# create csv's based on original training data

## avduarte333/BookTection

```python
import pandas as pd

df = pd.read_csv("hf://datasets/avduarte333/BookTection/0_BookTection_data.csv")
df
```

```python

df["ID"] = df["ID"].apply(lambda x: x.replace('_',' '))
df["ID"] = df["ID"].apply(lambda x: x.replace(' - ',' by '))
df
```

```python
df.Answer.unique()
```

```python
df.rename(columns={'Example_A': 'Text','ID':"Name",'Label':'Pre2022'}, inplace=True)
df
```

```python
df[['Name', 'Text','Pre2022','Length']].to_csv("books.csv",index=False)
```

### extract a sample of books to save cost 10 before 2022 and 10 after


```python
sample_pre_2022 = df[df['Pre2022'] == 1]['Name'].unique()[0:2] 
sample_after_2022 = df[df['Pre2022'] == 0]['Name'].unique()[0:2]

sample_after_2022
```

```python
df[df['Name'].isin(sample_pre_2022)][['Name', 'Text','Pre2022','Length']].to_csv("sample_books_pre_2022.csv",index=False)

df[df['Name'].isin(sample_after_2022)][['Name', 'Text','Pre2022','Length']].to_csv("sample_books_after_2022.csv",index=False)
```
