```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("result.csv")

score_columns = [col for col in df.columns if 'Score' in col]



```

```python
models = [' '.join(col.split("_")[1:-1]) for col in df.columns if 'Score' in col]
means = df[score_columns].mean()

sns.barplot(x=models, y=means.values, palette="Blues_d")
plt.title('Mean Recognition Scores Among Different Models')

plt.xticks(rotation=75)
plt.show()
```

```python
total_rows = len(df)
total_rows
```

```python
error_rate = df[score_columns].isna().sum() / total_rows
error_rate


sns.barplot(x=models, y=error_rate.values, palette="Blues_d")
plt.title('Mean error rate Among Different Models')

plt.xticks(rotation=75)
plt.show()
```
